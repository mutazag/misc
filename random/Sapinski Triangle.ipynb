```python
import random
from matplotlib import pyplot as plt

# create a list of three points in the shape of a triangle
points = [(0,0), (1,0), (0.5,0.5)]


# plot the points in red    
plt.plot(*zip(*points), marker='.', linestyle='none', color='red')
plt.show()
```

```python
#given triangle verticies, test if a point is inside the triangle
def isInsideTriangle(p, p0, p1, p2):
    #compute vectors
    v0 = [p2[0] - p0[0], p2[1] - p0[1]]
    v1 = [p1[0] - p0[0], p1[1] - p0[1]]
    v2 = [p[0] - p0[0], p[1] - p0[1]]
    
    # compute dot products
    dot00 = v0[0] * v0[0] + v0[1] * v0[1]
    dot01 = v0[0] * v1[0] + v0[1] * v1[1]
    dot02 = v0[0] * v2[0] + v0[1] * v2[1]
    dot11 = v1[0] * v1[0] + v1[1] * v1[1]
    dot12 = v1[0] * v2[0] + v1[1] * v2[1]
    
    #compute barycentric coordinates
    invDenom = 1 / (dot00 * dot11 - dot01 * dot01)
    u = (dot11 * dot02 - dot01 * dot12) * invDenom
    v = (dot00 * dot12 - dot01 * dot02) * invDenom
    
    #check if point is in triangle
    return (u >= 0) and (v >= 0) and (u + v < 1)




def generateRandomPoint():
    x = random.uniform(0, 1)
    y = random.uniform(0, 1)
    if x + y > 1:
        x = 1 - x
        y = 1 - y
    p = (x, y)
    return p


# generate random point until it is inside the triangle, keep a counter of number of trials
count = 0
while True:
    pspecial = generateRandomPoint()
    if isInsideTriangle(p, points[0], points[1], points[2]):
        break
    count += 1

print("Number of trials: ", count)

plt.plot(*zip(*points), marker='.', linestyle='none', color='red')
plt.plot(*zip(*[pspecial]), marker='.', linestyle='none', color='green')
plt.show()
```

```python

# func to return a midpoint between two points
def midpoint(p1, p2):
    return ((p1[0] + p2[0]) / 2, (p1[1] + p2[1]) / 2)


# func to return a random point from list 
def random_vertex(points):
    return random.choice(points)
```

```python
def generatePattern(points, p,  n):
    print(f'starting point: {p}')
    new_points = [p]
    for i in range(n):
        p1 = random_vertex(points)
        p = midpoint(p, p1)
        new_points.append(p)
    # plot vertecies in red and new points in green, vertecies are larger
    plt.plot(*zip(*points), marker='.', linestyle='none', color='red', markersize=10)
    plt.plot(*zip(*new_points), marker='.', linestyle='none', color='green', markersize=1)
    plt.show()


```

```python
generatePattern(points, pspecial, 50000)
```
